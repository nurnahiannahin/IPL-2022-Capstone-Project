1. Which team won the most matches?

```python
match_wins = ipl['match_winner'].value_counts()
sns.barplot(x = match_wins.values, y = match_wins.index, palette = 'viridis')
plt.xlabel('Number of Matchers')
plt.ylabel('Teams')
plt.title('Teams Won the Most Matches')
plt.show()
```

2. Toss Decision Trends.

```python
sns.countplot(x = ipl['toss_decision'])
plt.title('Toss Decision Trends')
plt.show()
```

3. Toss Winner vs Match Winner.

```python
count = ipl[ipl['toss_winner'] == ipl['match_winner']]['match_id'].count()
percentage = round((count/ipl['match_id'].count()) * 100)
percentage
```

4. How do teams win? (Runs or Wickets)

```python
sns.countplot(x = ipl['won_by'])
```

**Key Player Performance**

1. Most "Player of the match" Awards

```python
count = ipl['player_of_the_match'].value_counts().head(5)
count
```

```python
sns.barplot(x = count.values, y = count.index, palette = 'viridis')
plt.title("Top 5 Player of the Match Awards")
plt.show()
```

2 top Scorers

```python
high = ipl.groupby('top_scorer')['highscore'].sum().sort_values(ascending = False).head(2)
high
```

```python
high.plot(kind = 'barh')
```

10 Best Bowling Figures

```python
ipl.head(2)
```

```python
ipl['highest_wickets'] = ipl['best_bowling_figure'].apply(lambda x: x.split('--')[0])
ipl['highest_wickets'] = ipl['highest_wickets'].astype(int)
ipl
```

```python
high_wckt = ipl.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending = False).head(5)
high_wckt
```

```python
sns.barplot(x = high_wckt.values, y = high_wckt.index, palette = 'rocket')
plt.title('Top 5 Bowlers')
```

Most Matches Played by venue.
